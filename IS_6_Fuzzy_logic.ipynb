**Fuzzy Logic for Traffic light Control:**

Fuzzy logic is a computational paradigm that deals with uncertainty and imprecision by using linguistic variables, which are described by fuzzy sets. It&#39;s particularly useful in systems where traditional binary logic may not be as effective, as it allows for a more nuanced representation
of information. In the context of traffic light control, fuzzy logic can be employed to make decisions based on imprecise inputs, such as traffic density and waiting time.

Here's a theoretical overview of how fuzzy logic can be applied to traffic light control:
1. Input Variables:

- Traffic Density: This variable represents the amount of traffic on the road and can be categorized into fuzzy sets such as low, medium, and high.

- Waiting Time: This variable represents the time vehicles have been waiting at the intersection and can be categorized into fuzzy sets like short, medium, and long.

2. Output Variable:

- Light Phase: This variable represents the current state of the traffic light and can be divided into fuzzy sets such as red, yellow, and green.

3. Membership Functions:

- For each input and output variable, define membership functions that describe
the degree of membership of a value in a fuzzy set. These functions can be
triangular, trapezoidal, or other shapes depending on the desired characteristics.

4. Rules:
- Create a set of fuzzy rules that relate the input variables to the output variable.
-These rules express the decision-making logic of the traffic light control system.
For example:
- If traffic density is low and waiting time is short, then set the light phase to
green.
- If traffic density is medium or waiting time is medium, then set the light phase to
yellow.
- If traffic density is high or waiting time is long, then set the light phase to red.

5. Inference Engine:
- The inference engine evaluates the fuzzy rules based on the current input values and determines the appropriate fuzzy output.

6. Defuzzification:
- Convert the fuzzy output into a crisp value. This can be done using methods such as centroid, maximum membership, or other defuzzification techniques.

7. Control System:
- Combine all the components (input variables, membership functions, rules,
inference engine, and defuzzification) to create a fuzzy control system for the
traffic lights.
- By applying fuzzy logic in traffic light control, the system becomes capable of making decisions that consider the uncertainty and imprecision inherent in real-world traffic conditions. The flexibility of fuzzy logic allows for a more adaptive and responsive traffic light control system compared to traditional, rigid control systems.

#TRINAGULAR SHAPED FUNCTION

```python
def func_triangle(a,b,c):
    x=np.arange(0,100)
    y=[]
    for i in range(0,100):
        if x[i]<=a:
            y.append(0)
        elif a<=x[i]<=b:
            y.append((x[i]-a)/(b-a))
        elif b<=x[i]<=c:
            y.append((c-x[i])/(c-b))
        else:
            y.append(0)

    plt.plot(x,y)


func_triangle(5,18,40)
func_triangle(35,50,65)
func_triangle(55,75,95)
```

#TRAPEZOIDAL SHAPED FUNCTION

```python
def func_trapezoid(a,b,c,d):
    x=np.arange(0,100)
    y=[]
    for i in range(0,100):
        if x[i]<=a:
            y.append(0)
        elif a<=x[i]<=b:
            y.append((x[i]-a)/(b-a))
        elif b<=x[i]<=c:
            y.append(1)
        elif c<=x[i]<=d:
            y.append((d-x[i])/(d-c))
        else:
            y.append(0)

    plt.plot(x,y)


func_trapezoid(10,25,45,65)
func_trapezoid(20,35,50,70)
func_trapezoid(30,45,55,75)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
#Antecedent (input variable)
traffic_density= ctrl.Antecedent(np.arange(0,101,1), 'traffic_density')
waiting_time= ctrl.Antecedent(np.arange(0, 11, 1), 'waiting_time')
```

```python
#Consequent (output) variable
light_phase= ctrl.Consequent(np.arange(0,101,1), 'light_phase')
```

```python
#MembershipFunction
traffic_density['low']= fuzz.trimf(traffic_density.universe, [0,20,40])
traffic_density['medium']= fuzz.trimf(traffic_density.universe, [20,50,80])
traffic_density['high']= fuzz.trimf(traffic_density.universe, [60,80,100])
```

```python
traffic_density.view()
```

```python
waiting_time['short']= fuzz.trimf(waiting_time.universe, [0,2,4])
waiting_time['mid']= fuzz.trimf(waiting_time.universe, [2,5,8])
waiting_time['long']= fuzz.trimf(waiting_time.universe, [6,9,10])
```

```python
waiting_time.view()
```

```python
light_phase['red']= fuzz.trimf(light_phase.universe, [0,20,40])
light_phase['yellow']= fuzz.trimf(light_phase.universe, [20,50,80])
light_phase['green']= fuzz.trimf(light_phase.universe, [60,80,100])
```

```python
light_phase.view()
```

```python
#Define Rules
rule1= ctrl.Rule(traffic_density['low'] & waiting_time['short'], light_phase['green'])
rule2= ctrl.Rule(traffic_density['medium'] & waiting_time['mid'], light_phase['yellow'])
rule3= ctrl.Rule(traffic_density['high'] & waiting_time['long'], light_phase['red'])
```

```python
#Create a control system
traffic_light_ctrl= ctrl.ControlSystem([rule1, rule2, rule3])
```

```python
#Create a simulation
traffic_light= ctrl.ControlSystemSimulation(traffic_light_ctrl)
```

```python
#Set input values (eg. traffic density and waiting time)
traffic_light.input['traffic_density']=40
traffic_light.input['waiting_time']=6
```

```python
#Compute the result
traffic_light.compute()
```

```python
#Print the output value
print("Output light phase: ", traffic_light.output['light_phase'])
```

```python
#optimally plot the Membership Function and Simulation Result
light_phase.view(sim=traffic_light)
```
