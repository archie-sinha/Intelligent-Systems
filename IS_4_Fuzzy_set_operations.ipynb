#Fuzzy Sets:

Fuzzy refers to something that is unclear or vague. Hence, Fuzzy Set is a Set where every key is associated with value, which is between 0 to 1 based on the certainty. This value is often called as degree of membership.
Fuzzy logic is an organized method for dealing with imprecise data. It is a multivalued logic that allows intermediate values to be defined between conventional solutions. In classical set theory, the membership of elements in a set is assessed in binary terms according to a bivalent condition â€” an element either belongs or does not belong to the set. By contrast, fuzzy set theory permits the gradual assessment of the membership of elements in a set; this is described with the aid of a membership function valued in the real unit interval [0, 1].
Bivalent Set Theory can be somewhat limiting if we wish to describe a 'humanistic' problem mathematically. For example, Fig 1 below illustrates bivalent sets to characterize the temperature of a room. The most obvious limiting feature of bivalent sets that can be seen clearly from the diagram is that they are mutually exclusive - it is not possible to have membership of more than one set. Clearly, it is not accurate to define a transition from a quantity such as 'warm' to 'hot' by the application of one degree Fahrenheit of heat. In the real world a smooth (unnoticeable) drift from warm to hot would occur.

##Basic operations on Fuzzy Sets:

###Consider two fuzzy sets A and B.
1. Union:
Consider Y be the Union of them, then for every member of A and B, Y will be:

**Degree of membership(Y)= max(degree of membership(A), degree of membership(B))**

2. Intersection:
Consider Y be the Intersection of them, then for every member of A and B, Y will be:

**Degree of membership(Y)= min(degree of membership(A), degree of membership(B))**

3. Complement:
Consider Y be the Complement of it, then for every member of A, Y will be:

**Degree of membership(Y)= 1 â€“ degree of membership(A)**

4. Difference:  
Consider Y be the Intersection of them, then for every member of A and B, Y will be:
**Degree of membership(Y)= min(degree of membership(A), 1- degree of membership(B))**

```python
import numpy as np
```

```python
A = {"a": 0.4, "b": 0.2, "c": 0.7, "d": 0.3, "e": 0.5}
B = {"a": 0.7, "b": 0.6, "c": 0.3, "d": 0.5, "e": 0.2}
```

Union of Two Fuzzy Sets

```python
def union_dicts(A, B):
    union_AB = A.copy()
    union_AB.update(B)
    return union_AB


union_result = union_dicts(A, B)
print("Union Of sets is: ",union_result)
```

Intersection of Two Fuzzy Sets

```python
intersection = {key: A[key] for key in A.keys() & B.keys()}

print("Intersection of dictionaries A and B:", intersection)
```

Complement of Two Fuzzy Sets

```python
# Finding the complement of each probability
complement_A = {key: 1 - value for key, value in A.items()}
complement_B = {key: 1 - value for key, value in B.items()}

# Print the original and complement dictionaries
print("Original Dictionary A:", A)
print("Complement Dictionary A:", complement_A)
print("Original Dictionary B:", B)
print("Complement Dictionary B:", complement_B)
```

Difference between the Two Fuzzy Sets

```python
# Finding the difference between the values of corresponding keys
difference_result = {key: abs(A[key] - B[key]) for key in A}

# Print the original dictionaries and the difference result
print("Dictionary A:", A)
print("Dictionary B:", B)
print("Difference Result:", difference_result)
```

##Observations and Learning:
Fuzzy logic allows for gradual, overlapping truth values, accommodating degrees of uncertainty and imprecision, while traditional Boolean logic relies on crisp, binary true/false distinctions without considering degrees of truth. Fuzzy logic provides a more flexible framework for modeling real-world systems with ambiguous or incomplete information.

##Conclusion:

Fuzzy logic provides a flexible approach for handling uncertainty and imprecision in decision-making, allowing for nuanced reasoning and more realistic modeling in various fields. Its ability to manage vague boundaries enhances system adaptability and tolerance to incomplete information.
